```python
# read json
with open('speakers.json', encoding='utf-8') as f:
    speakers_data = json.loads(f.read())

# create dataframe
speakers_df = pd.json_normalize(speakers_data)

# save to csv
speakers_df.to_csv('speakers.csv', index=False, encoding='utf-8')
```
