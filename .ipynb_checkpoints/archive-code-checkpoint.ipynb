```python
corpus_df = pd.read_csv("movie-corpus/corpus.csv", header=None)
corpus_df_cleaned = pd.DataFrame(columns = ["Movie ID", "Text"])
corpus_df_cleaned["Movie ID"] = corpus_df.iloc[0]
corpus_df_cleaned["Text"] = corpus_df.iloc[1]
corpus_df_cleaned.drop(corpus_df_cleaned.tail(1).index, inplace = True)
display(corpus_df_cleaned)
```

```python
speakers_df_cleaned = speakers_df_cleaned.rename(columns={
    '_key':'Character ID',
    'meta/character_name':'Character Name',
    'meta/movie_idx':'Movie ID',
    'meta/movie_name':'Movie Name',
    'meta/gender':'Gender',
    'meta/credit_pos':'Credit Position'
})
speakers_df_cleaned.head()
```

```python
conversations_df = conversations_df.rename(columns={
    '_key':'Conversation ID',
    'meta/movie_idx':'Movie ID',
    'meta/movie_name':'Movie Name',
    'meta/release_year':'Release Year',
    'meta/rating':'Rating',
    'meta/votes':'IMBD Votes',
    'meta/genre':'Genre'
})
conversations_df.head()
```

```python
 # Model predicting gender by credit position
creditX = np.array(speakers_df_cleaned['Credit Position']).reshape(-1,1)
gender = np.array(speakers_df_cleaned['Gender'])

genderCredit = LogisticRegression().fit(creditX, gender)
print('credit coefficient:', genderCredit.coef_)
print('credit intercept:', genderCredit.intercept_)

credit_prediction = genderCredit.predict_proba(creditX)
credit_prediction = np.delete(credit_prediction, 1, 1)
speakers_df_cleaned['credit_prediction'] = credit_prediction
pyplot.show(seaborn.scatterplot(x=speakers_df_cleaned["Credit Position"], y=speakers_df_cleaned["Gender"], hue=speakers_df_cleaned["credit_prediction"], data="speakers_df_cleaned"))
```

**Data Description:**
Our data contains a list of movie conversations, information about their speakers, and information regarding the movie that contains the conversation. The columns are as follows:
- Character ID - A key value unique to each character
- Character Name - The name of the character speaking
- Movie ID - a key value unique to the movie in which the conversation takes place
- Movie Name - the name of the movie in which the conversation takes place
- Gender - the recorded gender of the speaker
- Credit Position - the position in the credits of the speaker's actor
- Release Year - the year of the movie's release
- Rating - the movie's rating on IMBD

**Data Limitations:**
As of right now, our dataset has a few limitations:
- The raw text of the conversation itself is not within the database
- The database may not be a perfect reflection of all movies over the time period
- The conversations may not represent all conversations for the movie they are taken from.

### Scraping
Scrape https://bechdeltest.com/?list=all to collect a list of movies, a binary value for whether or not they pass the test. This df will be merged with the original df containing relevant information of year, genre, and movie rating.

```python
import requests
from bs4 import BeautifulSoup
```

```python
list_url = 'https://bechdeltest.com/?list=all'
list_result = requests.get(list_url)
list_page = BeautifulSoup(list_result.text, 'html.parser')
```

```python
findMovies = list_page.find_all('div', {'class':'movie'})[:]

#make movie name array
movie_names = []
for x in findMovies:
    movie_names.append((x.text).replace('\n', '').lower())
print(len(movie_names))


#make pass/fail array
findPass = list_page.find_all('div', {'class':'list'})[0]
movie_pass = []
for x in findPass.find_all('img'):
    pass_fail = x['src']
    if pass_fail == '/static/pass.png':
            movie_pass.append(1)
    if pass_fail == '/static/nopass.png':
            movie_pass.append(0)
print(len(movie_pass))

#make dataframe with movie name + pass/fail metric
columns = ['Movie Name', 'Bechdel Pass/Fail']
bechdel_df = pd.DataFrame(columns=columns)
bechdel_df['Movie Name'] = movie_names
bechdel_df['Bechdel Pass/Fail'] = movie_pass
bechdel_df.head(n=10)
```
